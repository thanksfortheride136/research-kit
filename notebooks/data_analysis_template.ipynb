## Describe Data `.describe()`

This generate specific statistics about your dataset by column

- Count: Counts the number of non-null observations
- Mean: Mean of the values
- std: Standard deviation of the observations
- Min: Minimum values in the object
- Max: Maximum Values in the objet
- 25% (1st Quartile / Q1): The value where 25% of the data falls below it
- 50% (Median / Q2): The middle value of the dataset (half the data is below, half is above)
- 75% (3rd Quartile / Q3): The value where 75% of the data falls below it

You can specify specific colums using dot notation. For instance to display 1 column: ie. df.against_bug.describe(), two or more columns: df[["against_bug", "against_dragon"]].describe()

```python
df.describe()
```

```python
df[["against_bug", "against_dragon"]].describe()
```
