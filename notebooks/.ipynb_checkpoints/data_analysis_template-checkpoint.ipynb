## Describe Data `.describe()`

This generate specific statistics about your dataset by column

- Count: Counts the number of non-null observations
- Mean: Mean of the values
- std: Standard deviation of the observations
- Min: Minimum values in the object
- Max: Maximum Values in the objet
- 25% (1st Quartile / Q1): The value where 25% of the data falls below it
- 50% (Median / Q2): The middle value of the dataset (half the data is below, half is above)
- 75% (3rd Quartile / Q3): The value where 75% of the data falls below it

You can specify specific colums using dot notation. For instance to display 1 column: ie. df.against_bug.describe(), two or more columns: df[["against_bug", "against_dragon"]].describe()

```python
df.describe()
```

```python
df[["against_bug", "against_dragon"]].describe()
```

## Accessing and Selecting Data

Before cleaning or analyzing your data, you need to know how to access specific parts of your dataset.

This section covers:
- Selecting columns
- Selecting rows
- Filtering data using conditions

### Selecting Columns

You can select one or more columns from your dataset.

```python
# Select a single column
df["attack"]
```

```python
# Select multiple columns
df[["attack", "defense"]]
```

### Selecting Rows

You can select rows using their position (index).

```python
# First row
df.iloc[0]
```

```python
# First 5 rows
df.iloc[0:5]
```

### Selecting Rows and Columns Together

```python
# First row, first column
df.iloc[0, 0]
```

```python
# Rows 3-4, specific columns
df.iloc[3:5][["name", "attack", "defense"]]
```

### Filtering Data (Conditions)

You can filter your dataset based on certain conditions, this is one of the most important tools in data analysis.

```python
# Filter rows where attack is greater than 100
df[df["attack"] > 100]

```

```python
# Multiple conditions
df[(df["attack"] > 100) & (df["defense"] > 80) & (df["type1"] == "water")]
```

### Filtering Missing Data

You can use filtering to locate missing values in your dataset.

```python
# Rows where type2 is missing
df[df["type2"].isnull()]
```

### Sorting Data

You can sort your dataset to better understand trends.

```python
# Sort by attack (highest first)
df.sort_values(by="attack", ascending=False)
```

```python
# Sort by attack (highest first), notice I am using 2 brackets at the end to see 2 columns of info, making it clearer the name AND attack values. Also using head() to view 
#small amounts of the data at once.
df.sort_values(by="attack", ascending=False)[["name", "attack"]].head(20)
```
