# <span style="color:red">3. Inspect Your Data</span>

Use these tools to understand the structure, size, and content of your dataset before analyzing it.

### 📌 Quick Reference

**`df.head(n)`**  
Displays the first `n` rows

**`df.tail(n)`**  
Displays the last `n` rows

**`df.shape`**  
Returns dataset size as `(rows, columns)`

**`df.columns`**  
Displays all column names

**`df.info()`**  
Shows column names, data types, and non-null counts

**`df.describe()`**  
Shows summary statistics for numeric columns

**`df.isnull().sum()`**  
Shows how many missing values exist in each column

### 📌 Preview the Beginning of the Dataset with `df.head()`

**What this does:**  
Displays the first `N` rows of the dataset.

**Use when:**  
You want to quickly preview your data and see what the first records look like.

````python
df.head(5)
````

### 📌 Preview the End of the Dataset with `df.tail()`

**What this does:**  
Displays the last `N` rows of the dataset.

**Use when:**  
You want to inspect how the dataset ends or compare the beginning and end of the file.

````python
df.tail(5)
````

### 📌 Check Dataset Size with `df.shape`

**What this does:**  
Returns the size of the dataset as `(rows, columns)`.

**Use when:**  
You want to know how many observations and variables your dataset contains.

````python
print(f"Dataset Size → Rows: {df.shape[0]}, Columns: {df.shape[1]}")
````

### 📌 View All Column Names with `df.columns`

**What this does:**  
Displays the names of every column in the dataset.

**Use when:**  
You need to know what variables are available before selecting, filtering, or graphing data.

````python
print(df.columns)
````

### 📌 Check Dataset Structure with `df.info()`

**What this does:**  
Provides an overview of the dataset, including:
- number of rows and columns
- column names
- number of non-null values
- data types for each column

**Use when:**  
You want to understand the structure of the dataset and spot missing values or unexpected data types.

````python
df.info()
````

### 📌 Summary Statistics with `df.describe()`

**What this does:**  
Generates summary statistics for numeric columns.

**Common statistics included:**
- **count** → number of non-null values
- **mean** → average
- **std** → standard deviation
- **min** → minimum value
- **25%** → first quartile (Q1)
- **50%** → median (Q2)
- **75%** → third quartile (Q3)
- **max** → maximum value

**Use when:**  
You want a quick numerical summary of your dataset.

````python
df.describe()
````

### 📌 Describe Specific Columns

**What this does:**  
Generates summary statistics for selected columns only.

**Use when:**  
You only want to inspect a few variables instead of the entire dataset.

````python
df[["against_bug", "against_dragon"]].describe()
````

### 📌 Check for Missing Values with `df.isnull().sum()`

**What this does:**  
Counts how many missing values are in each column.

**Use when:**  
Before cleaning or analyzing the data.

⚠️ Missing data is very common in real-world datasets.

````python
df.isnull().sum()
````

# <span style="color:red">4. Select and Filter Data</span>

Most real data analysis involves selecting specific columns, rows, or subsets of data.

### 📌 Quick Reference

**Single column:**  
`df["column_name"]`

**Multiple columns:**  
`df[["col1", "col2"]]`

**Row by position:**  
`df.iloc[0]`

**Row range by position:**  
`df.iloc[3:5]`

**Filter rows:**  
`df[df["column"] > value]`

**Multiple conditions:**  
`df[(condition1) & (condition2)]`

**Sort values:**  
`df.sort_values(by="column", ascending=False)`

### 📌 Select One Column

**What this does:**  
Returns one column from the dataset.

**Use when:**  
You want to inspect or analyze one variable.

````python
df["attack"]
````

### 📌 Select Multiple Columns

**What this does:**  
Returns only the columns you specify.

**Use when:**  
You want to focus on a smaller set of variables.

````python
df[["attack", "defense"]]
````

### 📌 Select Rows by Position with `iloc`

**What this does:**  
Selects rows based on their index position.

**Use when:**  
You want to access specific rows directly.

⚠️ Python slicing is start-inclusive and end-exclusive.  
Example: `3:5` gives rows `3` and `4`, not `5`.

````python
# First row
df.iloc[0]
````

````python
# First 5 rows
df.iloc[0:5]
````

### 📌 Select Rows and Columns Together

**What this does:**  
Lets you target both specific rows and specific columns.

**Use when:**  
You want a smaller, more focused view of the dataset.

````python
# First row, first column
df.iloc[0, 0]
````

````python
# Rows 3-4, selected columns
df.iloc[3:5][["name", "attack", "defense"]]
````

### 📌 Filter Rows by a Condition

**What this does:**  
Returns only rows that meet a condition.

**Use when:**  
You want to analyze a subset of the data.

````python
df[df["attack"] > 100]
````

### 📌 Filter Using Multiple Conditions

**What this does:**  
Returns rows that satisfy more than one condition.

**Use when:**  
You want a more precise subset of the data.

⚠️ Each condition must be wrapped in parentheses.

````python
df[(df["attack"] > 100) & (df["defense"] > 80) & (df["type1"] == "water")]
````

### 📌 Filter Missing Data

**What this does:**  
Returns rows where a specific column is missing data.

**Use when:**  
You want to inspect incomplete records.

````python
df[df["type2"].isnull()] #Returns all rows where "type2" has no data
````

### 📌 Sort Data

**What this does:**  
Sorts the dataset by a selected column.

**Use when:**  
You want to find the highest or lowest values.

````python
df.sort_values(by="percentage_male", ascending=False)
````

### 📌 Sort + Select + Limit Output

**What this does:**  
Sorts the dataset, shows only the most useful columns, and limits the output.

**Use when:**  
You want a cleaner, easier-to-read result.

This is one of the most useful real-world patterns in pandas.

````python
df.sort_values(by="attack", ascending=False)[["name", "attack"]].head(20)
````

# <span style="color:red">5. Clean Your Data</span>

Before analyzing your data, you may need to clean it.

Common cleaning tasks include:
- handling missing values
- converting data types
- removing incorrect data

### 📌 Drop Rows with Missing Values

**What this does:**  
Creates a new DataFrame with rows containing missing values removed.

**Use when:**  
You want to work only with complete records.

⚠️ Be careful: dropping all missing rows can remove a large amount of data.

````python
df_clean = df.dropna()

print("Before:", df.shape)
print("After:", df_clean.shape)
````

### 📌 Convert a Column to Numeric

**What this does:**  
Attempts to convert a column into numbers.

**Use when:**  
A column should be numeric but is stored as text or mixed data. This will convert the text values to NaN, which allow you to use mean, average and other math methods on this. Otherwise you will throw errors.

`errors="coerce"` will turn invalid values into missing values (`NaN`).

Syntax `df["column_name"] = pd.to_numeric(df["column_name"], errors="coerce")`

````python
# Replace column_name with your actual column
df["against_fairy"] = pd.to_numeric(df["against_fairy"], errors="coerce")
````

### 📌 Recheck Missing Values After Cleaning

**What this does:**  
Lets you see whether your cleaning changed the number of missing values.

**Use when:**  
After converting data types or removing bad entries.

````python
df.isnull().sum()
````

# <span style="color:red">6. Visualize Your Data</span>

Use graphs to identify patterns, trends, and relationships.

### 📌 Quick Reference

**Histogram:**  
Good for showing the distribution of one numeric variable

**Scatter plot:**  
Good for showing the relationship between two numeric variables

**Box plot:**  
Good for comparing distributions across groups

**Count plot / bar chart:**  
Good for showing counts of categories

### 📌 Histogram

**What this does:**  
Shows the distribution of a numeric variable.

**Use when:**  
You want to see how values are spread out.

````python
df["attack"].hist()
plt.title("Attack Distribution")
plt.xlabel("Attack")
plt.ylabel("Frequency")
plt.show()
````

### 📌 Scatter Plot

**What this does:**  
Shows the relationship between two numeric variables.

**Use when:**  
You want to see whether variables appear related.

````python
sns.scatterplot(data=df, x="attack", y="defense")
plt.title("Attack vs Defense")
plt.show()
````

### 📌 Box Plot

**What this does:**  
Shows the spread of a numeric variable across categories.

**Use when:**  
You want to compare groups.

````python
print(df[["type1", "attack"]].dtypes)
print(df["type1"].isnull().sum(), df["attack"].isnull().sum())
print(df["type1"].nunique())
````

````python
plot_df = df[["type1", "attack"]].dropna().copy()

plt.figure(figsize=(12, 6))
sns.boxplot(data=plot_df, x="type1", y="attack")
plt.xticks(rotation=45)
plt.title("Attack by Primary Type")
plt.show()
````

### 📌 Count Plot

**What this does:**  
Shows how many observations fall into each category.

**Use when:**  
You want to compare category frequency.

````python
sns.countplot(data=df, x="type1")
plt.xticks(rotation=45)
plt.title("Count of Pokémon by Primary Type")
plt.show()
````
